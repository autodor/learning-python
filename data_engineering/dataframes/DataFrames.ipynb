#### Select specific columns from the dataset
Also specify that the decimal is a comma (European style) and not a dot (the norm outside Europe)

```python
required_cols = ['Country', 
                 'Population', 
                 'Area (sq. mi.)', 
                 'Birthrate', 
                 'Deathrate']
```

```python
countries = pd.read_csv('datasets/countries_of_the_world.csv', 
                        usecols = required_cols, 
                        decimal = ',')
```

```python
countries.head()
```

```python
countries['Country']
```

```python
countries['Country'].head()
```

#### Access specific element

```python
countries['Country'][206]
```

#### Slicing notation applies to Dataframes too
Pandas is built on top of NumPy so some operations overlap. The Dataframe is effectively a 2-D array with the column being the first dimension and the row is the second

```python
countries['Country'][105:112]
```

#### Select range of rows from specific columns
Note the array of string column names passed as opposed to the range of integer row indices

```python
countries[['Country', 'Birthrate']][105:112]
```

#### Selecting a range of columns is not possible with this notation

```python
countries['Country' : 'Birthrate'][105:112]
```

#### The loc operation
Allows access to specific rows and columns. This allows us to define a range of both rows and columns. 

Note that rows are listed first, then columns.

```python
countries.loc[105:112, 'Country':'Birthrate']
```

```python
countries.loc[105:112:3, 'Country':'Birthrate']
```

```python
countries.iloc[105:112, 0:4]
```
