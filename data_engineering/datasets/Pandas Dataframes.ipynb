```python
#ignore columns by passing in a list
required_cols = [
    'Country',
    'Population',
    'Area (sq. mi.)',
    'Birthrate',
    'Deathrate']
```

```python
#decimal = '.' to convert to period
countries = pd.read_csv('countries_of_the_world.csv',
                        usecols=required_cols,
                        decimal = '.'
                       )
countries.head()
```

```python
# access by index
countries['Country'][206]
```

```python
#return multiple by index
countries['Country'][105:112]
```

```python
#access multiple columns in a dataframe
#specify as a list
#restults in a dataframe (vs a series)
countries[['Country', 'Birthrate']][105:112]
```

```python
#to access a range of columns
#loc operator
countries.loc[105:112,'Country':'Birthrate']
```

```python
#more complex - step function. jumps by 3 after each result
countries.loc[105:112:3,'Country':'Birthrate']
```

```python
#instead of specifying by column name, specify columns by index id with iloc
countries.iloc[105:112,0:4]
```

```python

countries.iloc[105:112,4:12]
```
